```python
pd.read_html('https://www.fantasypros.com/nfl/reports/leaders/ppr.php?year=2022&start=1&end=18')
```

Best Ball ADP Data

```python
data = []

for year in range(2017, 2023):
    df_adp_year = pd.read_html(f'https://www.fantasypros.com/nfl/adp/best-ball-overall.php?year={year}')[0]

    df_adp_year.insert(0, 'Year', year)

    data.append(df_adp_year)

df_adp = pd.concat(data)

df_adp
```

```python
df_adp.drop(columns=['BB10', 'BB10', 'RTSports', 'Underdog', 'Drafters', 'AVG'], inplace=True)

# rename columns
df_adp.rename(columns={'Player Team (Bye)': 'Player', 'POS': 'Position', 'Rank': 'ADP'}, inplace=True)

# get rid of position ranking in number
df_adp['Position'] = df_adp['Position'].str.replace(r'\d+', '', regex=True)

# fix suspended, out, and IR
df_adp['Player'] = df_adp['Player'].str.replace(r' O$| IR$| S$', '', regex=True)

df_adp
```

```python
team_abbreviations = [
    "ARI", "ATL", "BAL", "BUF", "CAR", "CHI", "CIN", "CLE", "DAL", "DEN",
    "DET", "GB", "HOU", "IND", "JAC", "KC", "MIA", "MIN", "NE", "NO", "NYG",
    "NYJ", "LV", "PHI", "PIT", "LAC", "SF", "SEA", "LAR", "TB", "TEN", "WAS"
]

for team in team_abbreviations:
    df_adp['Player'] = [p[0] for p in df_adp['Player'].str.split(f' {team} ')]

df_adp
```

Weekly Scoring (PPR)

```python
data = []

for year in range(2017, 2023):
    df_score_year = pd.read_html(f'https://www.fantasypros.com/nfl/reports/leaders/ppr.php?year={year}&start=1&end=18')[0]

    df_score_year.insert(0, 'Year', year)

    data.append(df_score_year)

df_score = pd.concat(data)

df_score
```

```python
# only use 16 weeks
df_score.drop(columns=['#', 'Team', '17', 'AVG', 'TTL', '18'], inplace=True)

df_score.rename(columns={'Pos': 'Position'}, inplace=True)

for week in range(1, 17):
    df_score[f'{week}'] = pd.to_numeric(df_score[f'{week}'], errors='coerce')

df_score
```

```python
df = pd.merge(
    df_adp, df_score, 
    how='left', 
    on=['Year', 'Player', 'Position']
)

df[[f'{week}' for week in range(1, 17)]] = df[[f'{week}' for week in range(1, 17)]].fillna(0)

df
```

```python
df = df.loc[df['Position'].isin(['RB', 'WR', 'TE', 'QB']), :].reset_index(drop=True)

df
```

```python
df.loc[(df[[f'{week}' for week in range(1, 17)]].sum(axis=1) == 0) & (df['ADP'] < 100), :]
```

```python
df.to_csv('./data/adp_and_scoring.csv', index=False)

'Done'
```
